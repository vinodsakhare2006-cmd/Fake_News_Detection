# Fake News Detection\nThis notebook trains a fake news detection model using TF-IDF and Passive Aggressive Classifier.

```python
import pandas as pd\nimport pickle\nfrom sklearn.model_selection import train_test_split\nfrom sklearn.feature_extraction.text import TfidfVectorizer\nfrom sklearn.linear_model import PassiveAggressiveClassifier\nfrom sklearn.metrics import accuracy_score, classification_report, confusion_matrix\n
```

```python
df = pd.read_csv('../dataset/news.csv')\ndf.head()\n
```

```python
df['content'] = df['title'].fillna('') + ' ' + df['text'].fillna('')\nX_train, X_test, y_train, y_test = train_test_split(df['content'], df['label'], test_size=0.2, random_state=42, stratify=df['label'])\n
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.85, ngram_range=(1, 2))\nX_train_vec = vectorizer.fit_transform(X_train)\nX_test_vec = vectorizer.transform(X_test)\n
```

```python
model = PassiveAggressiveClassifier(max_iter=1000, random_state=42, class_weight='balanced')\nmodel.fit(X_train_vec, y_train)\npred = model.predict(X_test_vec)\nprint('Accuracy:', accuracy_score(y_test, pred))\nprint(classification_report(y_test, pred))\nprint(confusion_matrix(y_test, pred))\n
```

```python
pickle.dump(model, open('../model.pkl', 'wb'))\npickle.dump(vectorizer, open('../vectorizer.pkl', 'wb'))\nprint('Model and vectorizer saved')\n
```
